### Multiple line graph in one plot

```python
year = list(range(2007, 2016))
```

```python
print(year)
```

```python
Maruti = list(range(11,20))
Ford = list(range(31, 40))
Honda = list (range(41,50))
print(Maruti)
print(Ford)
print(Honda)
```

```python
plt.plot(year, Maruti , label = "Maruti", c = 'C0')
plt.plot(year, Ford , label = "Ford", c = 'C1')
plt.plot(year, Honda , label = "Honda", c = 'C2')
plt.legend()
plt.show()
```

#### Adding line style 

```python
plt.plot(year, Maruti , label = "Maruti", c = 'C0', ls = '--')
plt.plot(year, Ford , label = "Ford", c = 'C1', ls = '-.')
plt.plot(year, Honda , label = "Honda", c = 'C2', ls = ':')
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
xaxis = np.linspace (0, 200, 100) #it will generate 100 numbers in the range from 0 to 200
yaxis = xaxis**2

plt.plot(xaxis, yaxis, ls = "--", c = 'red', lw = '2', label = '$X^2$') # Legend label style changed by this
plt.legend()
plt.show()
```

### Draw plot using Marker 

```python
x = np.linspace(0, 20, 10) #it will give 10 numbers in the range of 0 to 20
print(x)
```

```python
plt.figure(figsize=(8,8)) # by this plot size can be change

line1 = x
line2 = x**2
line3 = x**2.1
line4 = x**2.2
line5 = x**2.3

plt.plot(x, line1, label='line1', marker = 's', lw= '2', ms= 10, c= 'red')
plt.plot(x, line2, label='line2', marker = '^', lw= '2', ms= 12, c= 'pink')
plt.plot(x, line3, label='line3', marker = 'o', lw= '2', ms= 8, c= 'blue')
plt.plot(x, line4, label='line4', marker = 'D', lw= '2', ms= 10, c= 'orange')
plt.plot(x, line5, label='line5', marker = 'P', lw= '2', ms= 12, c= 'purple')

plt.title('NonLinear lines', style = 'normal', size= 24)
plt.grid(color = 'black', ls= '-.', lw= '0.4')

plt.legend()
plt.show()
```

## Subplots 

```python
y = np.arange(200)
fig, plotnumbers = plt.subplots (2,2, sharex = False, sharey = False) 

# sharex & sharey false will define the xaxis & yaxis will be different for all four subplots.

plotnumbers[0,0].plot(-y)
plotnumbers[0,1].plot(y)
plotnumbers[1,0].plot(y**2)
plotnumbers[1,1].plot(-y*y)

plt.show()
```

```python
y = np.arange(200)
fig, plotnumbers = plt.subplots (2,2, sharex = False, sharey = False) 

# sharex & sharey false will define the xaxis & yaxis will be different for all four subplots.

plotnumbers[0,0].plot(-y)
plotnumbers[0,1].plot(y)
plotnumbers[1,0].plot(y**2)
plotnumbers[1,1].plot(-y*y)

plt.subplots_adjust(left =0, right = 2, top= 1.5, bottom = 0) # to adjust the subplots for better looking
plt.show()
```

```python
#other way to plot the four subplots

y = np.arange(10)

ax1 = plt.subplot(221) #subplot(221) means 2 rows, 2 column & plot no-1
plt.plot(-y*y)

ax2 = plt.subplot(222) #subplot(222) means 2 rows, 2 column & plot no-2
plt.plot(y*y)

ax3 = plt.subplot(223) #subplot(223) means 2 rows, 2 column & plot no-3
plt.plot(y*-y)

ax4 = plt.subplot(224) #subplot(224) means 2 rows, 2 column & plot no-4
plt.plot(-y*-y)

plt.subplots_adjust(left= 0, right= 1.5, top=1.5, bottom=0)
plt.show()
```

### Setting the inset plot in a big plot 

```python
x= np.linspace(0,20,40)
xcos = np.cos(x**2)
xsq = x**2

fig, ax1 = plt.subplots()

plt.subplots_adjust(left= 0, right=2, top=2, bottom=0)

left, bottom, width, height = [1.5, 1.6, 0.45, 0.35]
inset = fig.add_axes([left, bottom, width, height])

ax1.plot(x,xcos)
inset.plot(x, xsq, c='red', ls=':')
plt.show()
```
